```python
import kagglehub
import os
import pandas as pd

# Download latest version
path = kagglehub.dataset_download("argonnxx/teen-mental-health")

print("Path to dataset files:", path)

# List files in the downloaded path to identify the CSV file
print("Files in dataset directory:")
for dirname, _, filenames in os.walk(path):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
# Assuming the main dataset is 'teen_mental_health.csv' based on the dataset name
df = pd.read_csv(os.path.join(path, 'Teen_Mental_Health.csv'))

print("\nDataFrame loaded successfully. Here are the first 5 rows:")
display(df.head())
```

## Initial Data Inspection

Let's get a summary of the DataFrame to understand its structure, data types, and check for missing values.

```python
print("\nDataFrame Information:")
df.info()

print("\nDescriptive Statistics:")
display(df.describe(include='all'))
```

## Exploratory Data Analysis (EDA)

Let's begin by visualizing the distribution of key categorical variables and the overall prevalence.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set_style("whitegrid")

# Plotting distribution of categorical variables
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Distribution of Categorical Variables', fontsize=16)

sns.countplot(ax=axes[0], x='Gender', data=df, palette='viridis')
axes[0].set_title('Distribution of Gender')

sns.countplot(ax=axes[1], x='Age_Group', data=df, palette='magma')
axes[1].set_title('Distribution of Age Group')

sns.countplot(ax=axes[2], x='Mental_Health_Issue', data=df, palette='plasma')
axes[2].set_title('Distribution of Mental Health Issue')
axes[2].tick_params(axis='x', rotation=45)

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

### Prevalence Distribution

Now, let's look at the distribution of the `Prevalence` itself.

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Prevalence'], kde=True, bins=10, palette='cividis')
plt.title('Distribution of Prevalence')
plt.xlabel('Prevalence')
plt.ylabel('Frequency')
plt.show()
```

### Bivariate Analysis: Prevalence by Categorical Variables

Let's visualize the average prevalence for each category to understand their impact.

```python
fig, axes = plt.subplots(1, 3, figsize=(20, 6))
fig.suptitle('Average Prevalence by Categorical Variables', fontsize=18)

# Prevalence by Mental Health Issue
sns.barplot(ax=axes[0], x='Mental_Health_Issue', y='Prevalence', data=df, palette='viridis', ci=None)
axes[0].set_title('Average Prevalence by Mental Health Issue')
axes[0].tick_params(axis='x', rotation=45)

# Prevalence by Gender
sns.barplot(ax=axes[1], x='Gender', y='Prevalence', data=df, palette='magma', ci=None)
axes[1].set_title('Average Prevalence by Gender')

# Prevalence by Age Group
sns.barplot(ax=axes[2], x='Age_Group', y='Prevalence', data=df, palette='plasma', ci=None)
axes[2].set_title('Average Prevalence by Age Group')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

### Time Series Analysis: Prevalence Trends Over Years

Now, let's examine how the prevalence of mental health issues has changed over the years.

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Prevalence', data=df, hue='Mental_Health_Issue', marker='o', palette='tab10')
plt.title('Prevalence Trends of Mental Health Issues Over Years')
plt.xlabel('Year')
plt.ylabel('Prevalence')
plt.xticks(df['Year'].unique())
plt.legend(title='Mental Health Issue', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 7))
sns.lineplot(x='Year', y='Prevalence', data=df, hue='Age_Group', marker='o', palette='Dark2')
plt.title('Prevalence Trends by Age Group Over Years')
plt.xlabel('Year')
plt.ylabel('Prevalence')
plt.xticks(df['Year'].unique())
plt.legend(title='Age Group', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
